```python
def path_func (wire):
    '''function that converts a list of steps in the coordinates where a wire passes'''
    x=0
    y=0
    coordinates=[]
    for w in wire:
        s=0 #reset the number of steps given when reading a new instruction
        direction = w[0]
        steps = int(w[1:])
        while s < steps:
            if direction == 'R':
                x+=1
                s+=1
                coordinates.append([x,y])
            elif direction == 'L':
                x-=1
                s+=1
                coordinates.append([x,y])
            elif direction == 'U':
                y+=1
                s+=1
                coordinates.append([x,y])
            elif direction == 'D':
                y-=1
                s+=1
                coordinates.append([x,y])
            else:
                coordinates.append('invalid input')
                break
    return coordinates
        
```

```python
#determine wires' paths and intersections
wire1_path = path_func(wire1_list)
wire2_path = path_func(wire2_list)
intersections = []
for w1 in wire1_path:
    for w2 in wire2_path:
        if w1 == w2:
            intersections.append(w1)
            #break #makes the code more efficient (not tested)
        else:
            pass
print(intersections) 
  
```

```python
#determine the closest intersection to the central port
closest_intersection = 9999999999999999999999999999999999
for i in intersections:
    distance = abs(i[0])+abs(i[1])
    if distance <= closest_intersection:
        closest_intersection = distance
    else:
        pass
print(closest_intersection)
    
```

```python
def steps_func (wire):
    '''function that converts a list of steps in the coordinates where a wire passes as well as the number of steps taken to reach each coordinate'''
    x=0
    y=0
    total_steps = 0
    coordinates=[]
    for w in wire:
        s=0 #reset the number of steps given when reading a new instruction
        direction = w[0]
        steps = int(w[1:])
        while s < steps:
            if direction == 'R':
                x+=1
                s+=1
                total_steps += 1
                coordinates.append([[x,y],total_steps])
            elif direction == 'L':
                x-=1
                s+=1
                total_steps += 1
                coordinates.append([[x,y],total_steps])
            elif direction == 'U':
                y+=1
                s+=1
                total_steps += 1
                coordinates.append([[x,y],total_steps])
            elif direction == 'D':
                y-=1
                s+=1
                total_steps += 1
                coordinates.append([[x,y],total_steps])
            else:
                coordinates.append('invalid input')
                break
    return coordinates
```

```python
#orbit_map = ['COM)B','B)C','C)D','D)E','E)F','B)G','G)H','D)I','E)J','J)K','K)L']
f = open('day6_input.txt', "r")
orbit_map =  f.readlines()
#remove line break identifier and convert to int
orbit_map = [r.strip('\n') for r in orbit_map]

#transform inputs into list of strings
orbits = []
for i in orbit_map:
    orbits.append(i.split(")"))
predecessors = {o[1]:o[0] for o in orbits}
```

```python
def count_orbits(obj, predecessors):
    if obj not in predecessors:
        return 0  # Assume anything that doesn't appear is a common centre which orbits nothing
    else:
        return 1 + count_orbits(predecessors[obj], predecessors)

total_orbits = 0
for obj in predecessors:
    total_orbits += count_orbits(obj, predecessors)
    print(obj,str(count_orbits(obj, predecessors)))
    
print(total_orbits)
```

```python
#use graph theory library to calculate shortest path
import networkx as nx
orbit_map_G = nx.Graph() #initiate graph
for o in orbits:
    orbit_map_G.add_edge(o[1],o[0])

nx.shortest_path_length(orbit_map_G,[n for n in orbit_map_G.neighbors('YOU')][0],[n for n in orbit_map_G.neighbors('SAN')][0]) #neighbors returns an iterable dict key
```
